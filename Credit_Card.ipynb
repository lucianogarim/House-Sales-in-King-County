Análise Exploratória de Dados Cartão de Crédito

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Leitura e carregamento dos dados

```python
df = pd.read_csv('BankChurners.csv')
```

```python
df.head()
```

Vamos verificar se existem valores faltantes no nosso DataFrame

```python
df.isnull().values.any()
```

Retirando as colunas desnecessárias

```python
df=df.drop(columns=['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1'])
df=df.drop(columns=['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'])
df.head()
```

Uma breve descrição estatística de variáveis numéricas e categóricas

```python
df.describe() #Variáveis numéricas
```

```python
cat_cols = df.columns[df.dtypes == 'object']
df.describe(include=['O']) #Variáveis Categóricas
```

Avaliando a quantidade de clientes ativos e desligados

```python
df.groupby('Attrition_Flag').size().plot(kind='pie', autopct='%.2f')
```

Visualização de dados categóricos para os dois casos de clientes

```python
for col in cat_cols:
    g = sns.catplot(x=col,
                hue="Attrition_Flag",
                data=df, kind="count",
                height=6, aspect=.8)
plt.show()
```

Selecionando os clientes com problemas

```python
Att_Customer=df[df['Attrition_Flag'] == 'Attrited Customer']
```

```python
Att_Customer.head()
```

Vamos analisar o perfil desses clientes

```python
Att_Customer.describe(include=['O'])
```

```python
Att_Customer.describe()
```

```python
corr=Att_Customer.corr()
```

Vamos buscar por correlações entre os dados, usando um heatmap.

```python
plt.imshow(corr)
```

```python
s = corr.unstack()
so = s.sort_values(kind="quicksort")
so.tail(20)
```
