# Análise Exploratória de Dados em Python


O objetivo deste arquivo é fornecer os primeiros insights de um dataframe, a partir de uma análise exploratória nos dados.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Primeiramente, buscamos por um conjunto de dados disponibilizados pela IBM, e fazemos o carregamento dos dados.

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv'
df = pd.read_csv(url, header=None)
```

Inicialmente, é útil exibir uma pequena parte dos dados tabelados, e isto pode ser feito por meio da função head().

```python
df.head()
```

Note que o header do data frame é default, sendo assim, criaremos um header apropriado.

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns= headers 
df.head()
```

Para observarmos características gerais dos dados, podemos utilizar a função describe() para descrever as características estatísticas básicas dos dados. 

```python
df.describe()
```
